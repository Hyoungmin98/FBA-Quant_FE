```python
#(1)

def zero_coupon_bond_price(face_value, maturity_years, discount_rate):
    bond_price = face_value / ((1 + discount_rate) ** maturity_years)
    return bond_price

# 주어진 정보
face_value = 1000  # 액면가
maturity_years = 10  # 만기까지의 기간

# 각 채권의 가격을 계산
bond1_price = zero_coupon_bond_price(face_value, maturity_years, 0.04)
bond2_price = zero_coupon_bond_price(face_value, maturity_years, 0.04)
bond3_price = zero_coupon_bond_price(face_value, maturity_years, 0.08)

# 결과 출력
print("zero_coupon_bond_price1:", bond1_price)
print("zero_coupon_bond_price2:", bond2_price)
print("zero_coupon_bond_price3:", bond3_price)

# (b)  the price of a floating-rate bond = face value
print("Price of the floating-rate bond:", 1000)
```

```python
# 5
def count_ways_to_climb_stairs(n):
    if n <= 0:
        return 0
    # 1, 2, 3개의 개단을 오르는 경우
    elif n == 1:
        return 1
    elif n == 2:
        return 2
    elif n == 3:
        return 4
    # 
    else:
        return count_ways_to_climb_stairs(n-1) + count_ways_to_climb_stairs(n-2) + count_ways_to_climb_stairs(n-3)

num_stairs = 4
ways_to_climb = count_ways_to_climb_stairs(num_stairs)
print("경우의 수:", ways_to_climb)
```
