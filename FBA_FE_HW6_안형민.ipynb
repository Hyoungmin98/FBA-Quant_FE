```python
# 5
def moveZeroes(nums):
    non_zero_index = 0
    

    for i in range(len(nums)):
        if nums[i] != 0:
            nums[non_zero_index] = nums[i]
            non_zero_index += 1
    
    for i in range(non_zero_index, len(nums)):
        nums[i] = 0


nums1 = [0, 1, 0, 3, 12]
moveZeroes(nums1)
print(nums1)  
```

```python
# 6
def longestCommonPrefix(strs):
    if not strs:
        return ""
    
    prefix = strs[0]
    
    for s in strs[1:]:
        if not prefix:
            return ""
        
        i = 0
        while i < len(prefix) and i < len(s) and prefix[i] == s[i]:
            i += 1
        
        prefix = prefix[:i]
    
    return prefix

strs1 = ["flower", "flow", "flight"]
print(longestCommonPrefix(strs1))  
```
